```python
import torch
from ultralytics import YOLO

if torch.backends.mps.is_available():
    DEVICE = "mps"          # Apple Silicon GPU
elif torch.cuda.is_available():
    DEVICE = "0"            # first CUDA GPU
else:
    DEVICE = "cpu"
print("Using device:", DEVICE)
```

```python
DATA      = "dataset/data.yaml"   # from prepare_dataset.py
MODEL     = "yolo11n.pt"          # yolo11n (fast) / yolo11s / yolo11m (more capacity)
EPOCHS    = 200
IMGSZ     = 1024                  # drop to 640 if MPS runs low on memory
BATCH     = 4                     # lower if you hit OOM; -1 = auto
PROJECT   = "runs/chem"
NAME      = "yolo11n_chem"
PATIENCE  = 50                    # early-stop patience

import os
os.environ.setdefault("PYTORCH_ENABLE_MPS_FALLBACK", "1")  # safety net for unimplemented MPS ops

model = YOLO(MODEL)
model.train(
    data=DATA, epochs=EPOCHS, imgsz=IMGSZ, batch=BATCH, device=DEVICE,
    project=PROJECT, name=NAME, patience=PATIENCE,
    # augmentation: critical with a tiny dataset
    mosaic=1.0, close_mosaic=15, scale=0.5, translate=0.1,
    fliplr=0.5, flipud=0.0,        # no vertical flip -> keep text upright
    degrees=3.0,                   # pages are near-upright
    hsv_h=0.0, hsv_s=0.3, hsv_v=0.4,  # mostly brightness/contrast (ink is dark)
    mixup=0.0,
    # optimization
    optimizer="auto", lr0=0.01, warmup_epochs=5, cos_lr=True,
    # bookkeeping
    plots=True, seed=42, val=True,
)
```

```python
metrics = model.val(data=DATA, imgsz=IMGSZ, device=DEVICE)
print(f"mAP50-95: {metrics.box.map:.3f}")
print(f"mAP50   : {metrics.box.map50:.3f}")
for i, name in metrics.names.items():
    print(f"  {name:<20} AP50: {metrics.box.ap50[i]:.3f}")

print("\nBest weights:", f"{PROJECT}/{NAME}/weights/best.pt")
```

```python
#!/usr/bin/env python3
"""
Run olmOCR (served by your local vLLM, OpenAI-compatible API) over a folder of
reaction_scheme crops, using a chemistry-aware prompt.

Server assumed already running, e.g.:
    vllm serve allenai/olmOCR-2-7B-1025-FP8 --port 8000

Deps:
    pip install openai pyyaml pillow
"""
import base64, json, re
from io import BytesIO
from pathlib import Path

import yaml
from PIL import Image
from openai import OpenAI

# ------------------------- CONFIG -------------------------
SERVER       = "http://localhost:8000/v1"
API_KEY      = "EMPTY"                       # vLLM ignores it, but the client needs something
MODEL        = "allenai/olmOCR-2-7B-1025"  # must match the name vLLM is serving
CROPS_DIR    = "/home/rpx2985/yolo/runs/detect/predict-9/crops/reaction_scheme"
OUT_JSON     = "scheme_ocr.json"
LONGEST_DIM  = 1288    # 1288 for olmOCR-2 (1025); use 1024 for the 0225-preview
TEMPERATURE  = 0.1
MAX_TOKENS   = 1024
ROTATE_RETRY = True    # re-run once if the model says the crop is rotated
# ----------------------------------------------------------

client = OpenAI(base_url=SERVER, api_key=API_KEY)


def build_chem_scheme_prompt() -> str:
    return (
        "Attached is a cropped reaction scheme from a handwritten chemistry lab "
        "notebook: reactants, an arrow with conditions, and products.\n"
        "Transcribe what is WRITTEN. Distinguish two cases:\n"
        "- If a species is written as text or a chemical formula (e.g. EtONa, "
        " H_2O), transcribe that text exactly.\n"
        "- If a species is drawn as a molecular structure WITH a written name or label "
        "next to it, transcribe only that written label.\n"
        "- If a species is drawn as a structure with NO written name, do NOT name, "
        "guess, or describe what molecule it is. Mark it as [structure] in place and "
        "set is_diagram = true. Identifying drawn structures is out of scope.\n"
        "Also transcribe the arrow conditions exactly as written (solvent, temperature, "
        "catalyst, voltage, time, equivalents), e.g. A ->[conditions] B.\n"
        "Rules:\n"
        "- Do not hallucinate. Unreadable text -> [illegible]. Never substitute a "
        "plausible reagent for one you cannot actually read.\n"
        "- Preserve subscripts/superscripts/charges as LaTeX: \\(H_2O\\), \\(Li^+\\), "
        "\\(30^\\circ C\\). Use \\( \\) inline, \\[ \\] block; no unicode math symbols.\n"
        "- This is a fragment of a larger page; transcribe partial text as-is, do not "
        "complete cut-off words.\n"
        "- If there is no readable text at all, output null for natural_text.\n"
        "Return markdown with YAML front matter: primary_language, is_rotation_valid, "
        "rotation_correction, is_table, is_diagram."
    )


def encode_image(im: Image.Image, longest: int) -> str:
    im = im.convert("RGB")
    im.thumbnail((longest, longest))          # resize so longest dim == LONGEST_DIM
    buf = BytesIO()
    im.save(buf, format="PNG")
    return base64.b64encode(buf.getvalue()).decode()


def call_model(im: Image.Image) -> str:
    b64 = encode_image(im, LONGEST_DIM)
    resp = client.chat.completions.create(
        model=MODEL,
        temperature=TEMPERATURE,
        max_tokens=MAX_TOKENS,
        messages=[{"role": "user", "content": [
            {"type": "text", "text": build_chem_scheme_prompt()},
            {"type": "image_url", "image_url": {"url": f"data:image/png;base64,{b64}"}},
        ]}],
    )
    return resp.choices[0].message.content or ""


def parse_front_matter(text: str) -> tuple[dict, str]:
    """Split olmOCR's '---\\n<yaml>\\n---\\n<body>' into (meta_dict, body)."""
    m = re.match(r"\s*---\s*\n(.*?)\n---\s*\n?(.*)", text, re.DOTALL)
    if not m:
        return {}, text.strip()
    try:
        meta = yaml.safe_load(m.group(1)) or {}
    except yaml.YAMLError:
        meta = {}
    return meta, m.group(2).strip()


def ocr_crop(path: Path) -> dict:
    im = Image.open(path)
    raw = call_model(im)
    meta, body = parse_front_matter(raw)

    # One rotation retry if the model flags the crop as mis-oriented
    if (ROTATE_RETRY and meta.get("is_rotation_valid") is False
            and meta.get("rotation_correction") in (90, 180, 270)):
        # PIL rotates counter-clockwise; model reports clockwise correction -> negate
        im_rot = im.rotate(-int(meta["rotation_correction"]), expand=True)
        raw = call_model(im_rot)
        meta, body = parse_front_matter(raw)

    return {
        "file": path.name,
        "primary_language": meta.get("primary_language"),
        "is_rotation_valid": meta.get("is_rotation_valid"),
        "rotation_correction": meta.get("rotation_correction"),
        "is_diagram": meta.get("is_diagram"),
        "natural_text": None if body.strip().lower() == "null" else body,
        "raw": raw,
    }


def main():
    crops = sorted(p for p in Path(CROPS_DIR).iterdir()
                   if p.suffix.lower() in {".png", ".jpg", ".jpeg", ".webp"})
    if not crops:
        raise SystemExit(f"No crops found in {CROPS_DIR}")

    results = []
    for p in crops:
        try:
            r = ocr_crop(p)
            print(f"\n=== {p.name} ===")
            print(r["natural_text"])
            results.append(r)
        except Exception as e:
            print(f"[error] {p.name}: {e}")
            results.append({"file": p.name, "error": str(e)})

    Path(OUT_JSON).write_text(json.dumps(results, indent=2, ensure_ascii=False))
    print(f"\nWrote {len(results)} result(s) -> {OUT_JSON}")


if __name__ == "__main__":
    main()
```
